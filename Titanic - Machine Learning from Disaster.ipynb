```python
mean_value_test=test['Age'].mean()
test['Age'].fillna(value=mean_value_test, inplace=True)
test
test['Age'].apply(np.floor, inplace=True)
test.info()
```

```python
mean_value_Fare=test['Fare'].mean()
test['Fare'].fillna(value=mean_value_Fare, inplace=True)
test
test['Fare'].apply(np.floor, inplace=True)
test.info()
```

```python
test['Male'] = (test['Sex']=='male').astype('int')
test
```

                    Forecast -  Whether the passenger will survive or not?
                    Using Logistic Regresion.
                    Criterias of Regresion: 'Pclass', 'Male', 'Age', 'Fare' 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train = train[['Pclass', 'Male', 'Age', 'Fare' ]]
X_test = test[['Pclass', 'Male', 'Age', 'Fare' ]]
y_train = train['Survived']
y_test = gender_submission['Survived']
model = LogisticRegression()
model.fit(X_train, y_train)
y_predicted = model.predict(X_test)
y_predicted
```

```python
model.score(X_test, y_test)
#quality of regression“
```

```python
from sklearn import metrics
metrics.confusion_matrix(y_test, y_predicted)
#False negative 10
#False positive 8 
#True 256+144= 400
```
